```python
import numpy as np
import polars as pl
import polars.selectors as cs
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import great_tables as tg
import altair as alt
```

```python
df_path = r'../DATASETS/Data_cancer.csv'
```

```python
df = pl.read_csv(df_path)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.select(cs.exclude('Class'))
```

```python
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_regression = LogisticRegression()
```

```python
logistic_regression.fit(scaled_X_train, y_train)
```

```python
predictions = logistic_regression.predict(scaled_X_test)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
print(accuracy_score(y_test, predictions))
confusion_matrix(y_test, predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_classifier = KNeighborsClassifier()
```

```python
knn_classifier.fit(scaled_X_train, y_train)
```

```python
knn_predictions = knn_classifier.predict(scaled_X_test)
```

```python
print(accuracy_score(y_test, knn_predictions))
confusion_matrix(y_test, knn_predictions)
```

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(scaled_X_train, y_train)
```

```python
svc_predictions = svc_model.predict(scaled_X_test)
```

```python
print(accuracy_score(y_test, svc_predictions))
confusion_matrix(y_test, svc_predictions)
```

```python
svc_model_linear = SVC(kernel='linear')
```

```python
svc_model_linear.fit(scaled_X_train, y_train)
```

```python
svc_model_linear_predictions = svc_model_linear.predict(scaled_X_test)
```

```python
print(accuracy_score(y_test, svc_model_linear_predictions))
confusion_matrix(y_test, svc_model_linear_predictions)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gaussian_nb = GaussianNB()
```

```python
gaussian_nb.fit(scaled_X_train, y_train)
```

```python
gaussian_nb_predictions = gaussian_nb.predict(scaled_X_test)
```

```python
print(accuracy_score(y_test, gaussian_nb_predictions))
confusion_matrix(y_test, gaussian_nb_predictions)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decision_tree = DecisionTreeClassifier()
```

```python
decision_tree.fit(scaled_X_train, y_train)
```

```python
decision_tree_predictions = decision_tree.predict(scaled_X_test)
```
